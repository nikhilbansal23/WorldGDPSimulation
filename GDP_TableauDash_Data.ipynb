```python
# Drop the 'Code' column
gdp.drop(columns=['Code'], inplace=True, errors='ignore')

# Melt the DataFrame to reshape it into the desired schema
melted_gdp = pd.melt(gdp, id_vars=['Country Name'], var_name='Years', value_name='GDP')
# melted_gdp = melted_gdp.dropna()  # Remove rows with NaN values

# Sort the DataFrame based on 'Country Name' and 'Years' columns
melted_gdp = melted_gdp.sort_values(by=['Country Name', 'Years'])

# Display the resulting DataFrame
print(melted_gdp)
```

## **GDP PPP & Per Capita**

```python
# Drop the 'Code' column
gdp_ppp.drop(columns=['Code'], inplace=True, errors='ignore')

# Melt the DataFrame to reshape it into the desired schema
melted_gdp_PPP = pd.melt(gdp_ppp, id_vars=['Country Name'], var_name='Years', value_name='GDP_PPP')
# melted_gdp = melted_gdp.dropna()  # Remove rows with NaN values

# Sort the DataFrame based on 'Country Name' and 'Years' columns
melted_gdp_PPP = melted_gdp_PPP.sort_values(by=['Country Name', 'Years'])

# Display the resulting DataFrame
print(melted_gdp_PPP)
```

```python
# Drop the 'Code' column
gdp_ppp_per_capita.drop(columns=['Code'], inplace=True, errors='ignore')

# Melt the DataFrame to reshape it into the desired schema
melted_gdp_PPP_per_capita = pd.melt(gdp_ppp_per_capita, id_vars=['Country Name'], var_name='Years', value_name='GDP_PPP_percapita')
# melted_gdp = melted_gdp.dropna()  # Remove rows with NaN values

# Sort the DataFrame based on 'Country Name' and 'Years' columns
melted_gdp_PPP_per_capita = melted_gdp_PPP_per_capita.sort_values(by=['Country Name', 'Years'])

# Display the resulting DataFrame
print(melted_gdp_PPP_per_capita)
```

```python
melted_gdp_PPP_per_capita.to_csv('GDP PP PER CAPITA.csv',index=True)
```

## **Final Data**

```python
# Assuming df1, df2, df3, df4, df5, df6 are your DataFrames

# List of your DataFrames
dataframes = [melted_gdp, melted_gdp_growth, melted_gdp_per_capita, melted_gdp_per_capita_growth, melted_gdp_PPP, melted_gdp_PPP_per_capita]

# Ensure all DataFrames have the 'Country Name' and 'Years' columns
for df in dataframes:
    if 'Years' not in df.columns or 'Country Name' not in df.columns:
        raise ValueError("Missing 'Years' or 'Country Name' column in one or more DataFrames")

# Merge dataframes on common elements 'Country Name' and 'Years'
merged = dataframes[0]
for df in dataframes[1:]:
    merged = pd.merge(merged, df, on=['Country Name', 'Years'], how='outer')

# Remove duplicate columns, keeping the last occurrence
merged = merged.loc[:, ~merged.columns.duplicated()]

# Display the final merged DataFrame
print(merged)
```

```python
merged.to_csv('Full_GDP.csv',index=True)
```
